```python
# ============================================
# Big Tech Stocks — Financial Analysis
# Author: Marius Faur
# Tools: Python, Pandas, Matplotlib, Seaborn
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
sns.set_theme(style="whitegrid")
print("Libraries loaded successfully!")
```

```python
# Upload do ficheiro CSV
from google.colab import files
uploaded = files.upload()

# Carregar o dataset
df = pd.read_csv('15 Years Stock Data of NVDA AAPL MSFT GOOGL and AMZN.csv')

# Converter Date para datetime
df['Date'] = pd.to_datetime(df['Date'])

# Primeira exploração
print(f"Dataset shape: {df.shape}")
print(f"\nPeriod: {df['Date'].min()} to {df['Date'].max()}")
print(f"\nColumns: {df.columns.tolist()}")
print(f"\nMissing values:\n{df.isnull().sum()}")
df.head()
```

```python
# Calcular retorno anual por empresa
df['Year'] = df['Date'].dt.year

companies = {
    'AAPL': 'Close_AAPL',
    'AMZN': 'Close_AMZN',
    'GOOGL': 'Close_GOOGL',
    'MSFT': 'Close_MSFT',
    'NVDA': 'Close_NVDA'
}

# Preço de fecho no início e fim de cada ano
annual_returns = {}
for ticker, col in companies.items():
    yearly = df.groupby('Year')[col].agg(['first', 'last'])
    yearly['return_pct'] = ((yearly['last'] - yearly['first']) / yearly['first']) * 100
    annual_returns[ticker] = yearly['return_pct']

returns_df = pd.DataFrame(annual_returns)

# Visualização
fig, ax = plt.subplots(figsize=(16, 7))
returns_df.plot(kind='bar', ax=ax, colormap='tab10')
ax.set_title('Annual Return by Company (%)', fontsize=14, fontweight='bold')
ax.set_xlabel('Year')
ax.set_ylabel('Annual Return (%)')
ax.axhline(y=0, color='black', linewidth=0.8, linestyle='--')
ax.legend(loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('annual_returns.png', dpi=150, bbox_inches='tight')
plt.show()
print(returns_df.round(1))
```

```python
# Correlação entre os preços de fecho das 5 empresas
close_cols = ['Close_AAPL', 'Close_AMZN', 'Close_GOOGL', 'Close_MSFT', 'Close_NVDA']
corr_matrix = df[close_cols].corr()

# Renomear para melhor leitura
corr_matrix.columns = ['AAPL', 'AMZN', 'GOOGL', 'MSFT', 'NVDA']
corr_matrix.index = ['AAPL', 'AMZN', 'GOOGL', 'MSFT', 'NVDA']

fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='Blues',
            ax=ax, linewidths=0.5, vmin=0, vmax=1)
ax.set_title('Stock Price Correlation Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('correlation_matrix.png', dpi=150, bbox_inches='tight')
plt.show()
print("\nCorrelation Matrix:")
print(corr_matrix.round(2))

plt.savefig('best_worst_years.png', dpi=150, bbox_inches='tight')
print("Saved!")
```

```python
# Sumário final de insights
print("=" * 55)
print("KEY INSIGHTS — Big Tech Stock Analysis (2010–2024)")
print("=" * 55)

for company, col in companies.items():
    first_price = df[col].iloc[0]
    last_price = df[col].iloc[-1]
    total_return = ((last_price - first_price) / first_price) * 100
    best_year = returns_df[company].idxmax()
    worst_year = returns_df[company].idxmin()
    print(f"\n{company}:")
    print(f"  Price 2010: ${first_price:.2f} → 2024: ${last_price:.2f}")
    print(f"  Total return: {total_return:.0f}%")
    print(f"  Best year: {best_year} ({returns_df.loc[best_year, company]:.0f}%)")
    print(f"  Worst year: {worst_year} ({returns_df.loc[worst_year, company]:.0f}%)")

print("\nAnalysis complete!")
```
